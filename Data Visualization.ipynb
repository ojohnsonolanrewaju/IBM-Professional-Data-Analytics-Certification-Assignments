## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*) as No_of_Rows
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Bubble Plots


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)
df
```

Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# WorkWeekHrs
ax0 = df.plot(kind='scatter',
                    x='Age',
                    y='WorkWeekHrs',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s='Age'
                    )

# CodeRevHrs
ax1 = df.plot(kind='scatter',
                    x='Age',
                    y='CodeRevHrs',
                    alpha=0.5,
                    color="blue",
                    s='Age',
                    ax=ax0
                    )
ax0.set_ylabel('Number of Working Hrs')
ax0.set_title('Relationship between Hours and Age')
ax0.legend(['WorkWeekHrs', 'CodeRevHrs'], loc='upper left', fontsize='x-large')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
LanguageDesireNextYear
```

```python
QUERY = """
SELECT count(LanguageDesireNextYear) as Count,LanguageDesireNextYear
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count(LanguageDesireNextYear) 
DESC LIMIT 5
"""
df= pd.read_sql_query(QUERY,conn)
df
```

```python
QUERY = """ select LanguageWorkedWith, count(Respondent) as count   
from LanguageWorkedWith  
where LanguageWorkedWith = 'SQL'
"""
df= pd.read_sql_query(QUERY,conn)
df
```

```python
QUERY = """

select count(DatabaseDesireNextYear) as Count,DatabaseDesireNextYear 
from DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count(DatabaseDesireNextYear) 
DESC LIMIT 5
"""
df= pd.read_sql_query(QUERY,conn)
df
```

```python
explode_list = [0.0, 0, 0, 0.1, 0.1]
```

```python
# autopct create %, start angle represent starting point
df['Count'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Top 5 databases that respondents wish to learn next year', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df.DatabaseDesireNextYear, loc='upper left') 

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age >= 30 AND Age < 35
"""
df= pd.read_sql_query(QUERY,conn)
df
```

```python
order = ['WorkWeekHrs', 'CodeRevHrs']
df.groupby('Age')[order].median().plot.bar(stacked=True)
plt.title('Median hours, by age')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 AND Age < 60
"""
df= pd.read_sql_query(QUERY,conn)
df
```

```python
Converted = ['ConvertedComp']
```

```python
df1 = df.groupby('Age')[Converted].median()
df1
```

```python
df1.plot(kind='line',figsize=(10, 6))
plt.title('Median ConvertedComp for all ages from 45 to 60.')
plt.ylabel('Median ConvertedComp')
plt.xlabel('Age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch
FROM master
WHERE MainBranch IS NOT NULL
"""

bar_data = pd.read_sql_query(QUERY, conn)
bar_data.value_counts().plot(
    kind='barh',
    figsize=(5, 5),
    rot=360
)
plt.ylabel('Sentiment')
plt.xlabel('Number of Respondents ')
plt.title('Main Branch at Work')

plt.show()
```

```python
QUERY_b = """
SELECT count(MainBranch) as Count, MainBranch 
FROM master
group by MainBranch
"""
df_b=pd.read_sql_query(QUERY_b,conn)
df_b.set_index('MainBranch',inplace=True)
```

```python
df_b.plot(kind='barh', figsize=(7, 5), color='steelblue')
plt.xlabel('Numbers of Respondents')
plt.title('Distribution of Main Branch at Work')
```
